## Multiple soil levels and surface data

```python
import earthkit.data

fname = "soil_and_sfc_levels.grib"
earthkit.data.download_example_file(fname, remote_dir="test-data/xr_engine/level")
ds_ek = earthkit.data.from_source("file", fname)
```

```python
ds = ds_ek.to_xarray()
ds
```

```python
ds = ds_ek.to_xarray(profile="cf")
ds
```

```python
ds = ds_ek.to_xarray(profile="cf", ensure_dims=["level"], dims_as_attrs="level")
ds
```

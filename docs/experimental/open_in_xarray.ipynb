## Opening in xarray

```python
import earthkit.data
import xarray as xr

fname = "pl_regular_ll.grib"
earthkit.data.download_example_file(fname, remote_dir="test-data/xr_engine/level")
#ds_ek = earthkit.data.from_source("file", fname)
```

#### Using filepath

```python
xr.open_dataset(fname, engine="earthkit", time_dim_mode="raw")
```

```python
xr.open_dataset(fname, engine="earthkit", backend_kwargs={"time_dim_mode": "raw"})
```

#### Using earthkit object

```python
ds_ek = earthkit.data.from_source("file", fname)
xr.open_dataset(ds_ek, engine="earthkit", time_dim_mode="raw")
```

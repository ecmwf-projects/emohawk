## Convert Climate DT data into CF compliant NetCDF

```python
import earthkit.data

r = {
        "class": "d1",
        "dataset": "climate-dt",
        "activity": "ScenarioMIP",
        "experiment": "SSP3-7.0",
        "generation": "1",
        "realization": "1",
        "model": "IFS-NEMO",
        "resolution": "high",
        "expver": "0001",
        "stream": "clte",
        "date": "20220903/20220904", # 20200101/20200102
        "time": "0000/0012", # 1/to/12/by/1
        "step": "0/12",
        "type": "fc",
        "levtype": "hl",
        "levelist" : "100",
        "param": "228246/228247",
    }
 
ds = earthkit.data.from_source("polytope", "destination-earth", r, 
                                 address="polytope.lumi.apps.dte.destination-earth.eu", 
                                 stream=False)

kwargs = {"xarray_open_dataset_kwargs": {
            "backend_kwargs": {
                "profile": "cf",
                "squeeze": True                
            }
          }
         }


a = ds.to_xarray(**kwargs)
a
```

```python
a.to_netcdf("_climate_dt_healpix.nc")
```

```python
import subprocess
result = subprocess.run(["ncdump", "-h", "_climate_dt_healpix.nc"], stdout=subprocess.PIPE)
t = result.stdout.decode('utf-8').split('\n')
for line in t:
    print(line)
```

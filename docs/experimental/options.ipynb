## Options

```python
import earthkit.data

fname = "pl_regular_ll.grib"
earthkit.data.download_example_file(fname, remote_dir="test-data/xr_engine/level")
ds_ek = earthkit.data.from_source("file", fname)
```

```python
ds = ds_ek.to_xarray()
ds
```

```python
ds = ds_ek.to_xarray(time_dim_mode="raw", decode_time=False, 
                     add_valid_time_coord=True)
ds
```

```python
ds = ds_ek.to_xarray(time_dim_mode="valid_time", decode_time=True)
ds
```

```python
ds = ds_ek.to_xarray(level_dim_mode="level_per_type")
ds
```

```python
ds = ds_ek.to_xarray(level_dim_mode="level_and_type")
ds
```

```python
ds = ds_ek.to_xarray(variable_key="param_level")
ds
```

```python
ds = ds_ek.to_xarray(variable_key="param_level_type", 
                     remapping={"param_level_type": "{param}_{level}_{levtype}"})
ds
```

```python
ds = ds_ek.to_xarray(extra_dims="type")
ds
```

```python
ds = ds_ek.to_xarray(ensure_dims="type")
ds
```

```python
ds = ds_ek.to_xarray(drop_variables="r")
ds
```

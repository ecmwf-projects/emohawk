## Seasonal forecast

```python
import earthkit.data

fname = "jma-seasonal-forecast-forecast-ref-time-per-member.grib"
earthkit.data.download_example_file(fname, remote_dir="test-data/xr_engine/date")
ds_ek = earthkit.data.from_source("file", fname)
```

```python
ds = ds_ek.sel(param=["2t"]).to_xarray(
    profile="cf", 
    time_dim_mode="forecast", 
    time_dim_mapping={"date": "indexingDate", "time": "indexingTime", "step": "forecastMonth"})

ds
```

```python
ds = ds_ek.sel(param=["2t"]).to_xarray(
    profile="cf", 
    time_dim_mode="forecast", 
    time_dim_mapping="seasonal")

ds
```

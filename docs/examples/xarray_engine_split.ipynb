## Xarray engine

```python
from earthkit.data import from_source
from earthkit.data.testing import earthkit_remote_test_data_file

ds1 = from_source("url", earthkit_remote_test_data_file("test-data", "xr_engine", "level", "pl_regular_ll.grib"))
ds2 = from_source("url", earthkit_remote_test_data_file("test-data", "xr_engine", "level", "ml_regular_ll.grib"))
ds1 = ds1.sel(param=["t", "u", "v"])
ds = ds1 + ds2
```

```python
#ds.ls()
```

### Using cfgrib

```python
try:
    a = ds.to_xarray(_legacy=True)
except Exception as e:
    print(e)
```

### Using earthkit engine

#### Auto split

```python
kwargs = {"xarray_open_dataset_kwargs": {
            "backend_kwargs": {
                "profile": "mars", 
                "auto_split": True
        }
      }}

a = ds.to_xarray(**kwargs)
len(a)
```

## Xarray engine: level options

#### level_dim_mode=level

```python
import earthkit.data as ekd
ds_fl = ekd.from_source("sample", "pl.grib")
ds_xr = ds_fl.to_xarray()
ds_xr
```

```python
ds_xr = ds_fl.to_xarray(profile="grib")
ds_xr
```

#### level_dim_mode=level_and_type

When ``level_dim_mode="level_and_type"`` a single dimension called "level_and_type" will be formed by combining the keys assigned to the level and level_type roles. This option can be used to handle data with mutiple level types for the same parameters. The GRIB data in the next example contains both pressure and hybrid (model) level data.

```python
ds_fl = ekd.from_source("sample", "mixed_pl_ml.grib")
ds_xr = ds_fl.to_xarray(level_dim_mode="level_and_type")
ds_xr
```

#### level_dim_mode=level_per_type

```python
ds_fl = ekd.from_source("sample", "mixed_pl_sfc.grib")
ds_xr = ds_fl.to_xarray(profile="grib", level_dim_mode="level_per_type", squeeze=False)
ds_xr
```

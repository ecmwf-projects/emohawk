## Xarray engine

```python
import earthkit.data
earthkit.data.download_example_file("pl_regular_ll.grib", remote_dir="test-data/xr_engine/level")

# pressure level data on regular_ll grid
ds = earthkit.data.from_source("file", "pl_regular_ll.grib")
```

### Using cfgrib

```python
a = ds.to_xarray(_legacy=True)
a
```

### Using earthkit engine

#### MARS keys only

```python
kwargs = {"xarray_open_dataset_kwargs": {
            "backend_kwargs": {
                "profile": "mars", 
        }
      }}

a = ds.to_xarray(**kwargs)
a
```

#### Merge date and time into base_datetime

```python
kwargs = {"xarray_open_dataset_kwargs": {
            "backend_kwargs": {
                "profile": "mars", 
                "base_datetime_dim": True
        }
      }}

a = ds.to_xarray(**kwargs)
a
```

#### Add valid datetime coord and use timedelta for steps

```python
kwargs = {"xarray_open_dataset_kwargs": {
            "backend_kwargs": {
                "profile": "mars", 
                "base_datetime_dim": True,
                "valid_datetime_coord": True,
                "timedelta_step": True
                
        }
      }}

a = ds.to_xarray(**kwargs)
a
```

#### Flatten values

```python
kwargs = {"xarray_open_dataset_kwargs": {
            "backend_kwargs": {
                "profile": "mars", 
                "flatten_values": True
        }
      }}

a = ds.to_xarray(**kwargs)
a
```

#### Use separate level dimension per level type

```python
kwargs = {"xarray_open_dataset_kwargs": {
            "backend_kwargs": {
                "profile": "mars", 
                "level_per_type_dim": True
        }
      }}

a = ds.to_xarray(**kwargs)
a
```

#### Use remappings

```python
kwargs = {"xarray_open_dataset_kwargs": {
            "backend_kwargs": {
                "profile": "mars", 
                "remapping": {"param_lev": "{param}_{levelist}_{levtype}"}
        }
      }}

a = ds.to_xarray(**kwargs)
a
```

#### Built-in remappings

```python
kwargs = {"xarray_open_dataset_kwargs": {
            "backend_kwargs": {
                "profile": "mars", 
                "extra_index_keys": ["param_level"]
        }
      }}

a = ds.to_xarray(**kwargs)
a
```

```python
kwargs = {"xarray_open_dataset_kwargs": {
            "backend_kwargs": {
                "profile": "mars", 
                "extra_index_keys": ["level_and_type"]
        }
      }}

a = ds.to_xarray(**kwargs)
a
```
